# Temel Tanımlayıcı Bilgiler:
Accident_Index:
Her kaza için benzersiz bir ID.

Location_Easting_OSGR, Location_Northing_OSGR:
İngiltere’nin kendi koordinat sistemine göre (OSGR) doğu ve kuzey konumları.

Longitude, Latitude:
GPS koordinatları. Haritada direkt konum göstermek istersen bunu kullanırsın.  

# Olayla İlgili Bilgiler:
Police_Force:
Olayı raporlayan polis teşkilatının ID’si. Belki kurumlar arası performans karşılaştırmasında kullanılabilir.

Accident_Severity:
Kazanın ciddiyeti (muhtemelen 1 = Fatal, 2 = Serious, 3 = Slight).
🎯 Bu hedef değişken olabilir! (sınıflandırma problemi)

Number_of_Vehicles:
Kazaya karışan araç sayısı.

Number_of_Casualties:
Yaralanan veya ölen kişi sayısı. Özellikle şiddetli kazaları anlamada önemli.

# Zaman Bilgileri:
Date:
Kazanın tarihi. YYYY-MM-DD formatında olur genelde.
→ Yıl, ay, hafta, hafta içi vb. çıkarılabilir.

Day_of_Week:
Pazartesi – Pazar arası gün. Hafta sonları daha çok kaza oluyor mu diye bakabilirsin.

Time:
Saat bilgisi (HH:MM). Yoğun saatlerde kaza artıyor mu? Pik saat analizi yapılabilir.

# Yerel Bilgiler:
Local_Authority_(District):
Kazanın gerçekleştiği yerel yönetim (ilçe).

Local_Authority_(Highway):
Yolun bakımından sorumlu olan birim. Yolların kalitesiyle ilişki kurulabilir.

# Yol Bilgileri:
1st_Road_Class, 1st_Road_Number:
Kaza yerindeki ana yolun sınıfı ve numarası.
Örnek: A sınıfı ana yol, B tali yol gibi olabilir.

Road_Type:
Tek yön mü, çift yön mü, kavşak mı vs.

Speed_limit:
O yolun yasal hız sınırı (km/saat). Hız – şiddet ilişkisi analiz edilebilir.

Junction_Detail, Junction_Control:
Kaza bir kavşakta mı olmuş? Işık, levha var mıydı?
Işık olan kavşaklarda kaza sayısı daha az mı gibi çıkarım yapılabilir.

2nd_Road_Class, 2nd_Road_Number:
Eğer kavşak varsa ikinci yolla ilgili bilgiler.

# Yaya & Fiziksel Koşullar:
Pedestrian_Crossing-Human_Control:
Görevli bir yaya geçidi kontrolü (örneğin trafik polisi).
Yes/No olabilir.

Pedestrian_Crossing-Physical_Facilities:
Fiziksel geçit var mı? (zebra çizgisi, alt geçit vs.)

# Ortam Koşulları:
Light_Conditions:
Gündüz, gece, sokak lambalı gece gibi seçenekler olabilir.
→ Görüş etkisi.

Weather_Conditions:
Yağmur, sis, kar, açık hava gibi seçenekler.
🌧️ Kötü havalarda kaza artıyor mu?

Road_Surface_Conditions:
Yolun ıslak mı kuru mu olduğu.
Su birikintileri, buzlanma gibi şeyler olabilir.

Special_Conditions_at_Site:
Yol çalışması, yol daralması gibi özel durumlar.

Carriageway_Hazards:
Yolda engel, hayvan, döküntü var mı?

# Lokasyon Özellikleri:
Urban_or_Rural_Area:
Şehir içi mi, kırsal mı?
Şehirde daha sık ama kırsalda daha ölümcül kazalar olabilir.

Did_Police_Officer_Attend_Scene_of_Accident:
Olay yerine polis gelmiş mi? İlginç bir korelasyon yakalanabilir.

LSOA_of_Accident_Location:
İngiltere’deki en küçük bölgesel istatistik birimi (Lower Layer Super Output Area).
Detaylı coğrafi analiz için kullanılabilir.

# Genel:
Year:
Kazanın gerçekleştiği yıl. Yıllara göre artış var mı? COVID sonrası değişim? vs.

- Location_Easting_OSGR: değişkeni, ortalama ve medyanın birbirine oldukça yakın olmasıyla simetrik bir dağılıma sahip gibi görünüyor.
Q1–Q3 arası çeyrek değerler dengeli dağıldığı için, verinin büyük bir kısmı orta bölgede toplanmış durumda.
Ayrıca standart sapma, ortalamanın yaklaşık %21’i kadar olduğundan, veri çok yaygın değil; ılımlı bir yayılım gösteriyor.
Genel olarak bu değişken, temiz ve dengeli bir sayısal dağılıma sahip diyebiliriz.

- Location_Northing_OSGR: Easting ile benzer dağılıma sahip.

- Longitude: simetrik dağılım, dağılım ortada toplanmış.
- Latitude: simetrik dağılım, dağılım ortada toplanmış ve std meanden oldukça küçük olduğu için veri dar bir aralıkta yoğunlaşmış.
- Police_Force: Sayısal gibi gözükse de aslında her biri bir polis birimini temsil eden kategorik ID'dir. Ortalama (30.2) ve medyan (30) birbirine yakın → yaklaşık simetrik. Max değeri 98 → uç değer olabilir, gerçekçi olup olmadığı kontrol edilmeli. Sayı gibi analiz edilmemeli, kategori olarak işlenmeli.
- Accident_Severity: 1 (ölümcül), 2 (ciddi), 3 (hafif) olarak kodlanmış ordinal kategorik değişken. Medyan ve tüm çeyrek değerler 3 → veri %75’ten fazlası hafif kaza. Aşırı dengesiz dağılım (imbalanced) → sınıflandırmada SMOTE, class_weight gibi yöntemlerle ele alınmalı.
- Number_of_Vehicles: Sürekli sayısal bir değişken. Medyan 2, ortalama 1.83 → verilerin çoğu 1-2 araçlı kaza. Max 67 → uç değerler var, dağılım pozitif çarpık. İyileştirme için log-transform gibi yöntemler düşünülebilir.
- Number_of_Casualties: Sürekli sayısal bir değişken. Medyan 1, ortalama 1.35 → çoğu kazada 1 kişi yaralanmış. Max 93 → ciddi uç değer var, dağılım aşırı pozitif çarpık.
- Day_of_Week: 1–7 arasında değer alan kategorik (ordinal) değişken. Ortalama 4.11 → dağılımın hafif ortalarda yoğunlaştığını gösterir. Sayı gibi işlenmemeli, kategori olarak ele alınmalı. En doğru analiz, bar plot ya da countplot ile dağılım görselleştirmesi olur.
- Local_Authority_(District): Sayı gibi görünse de aslında her biri bir yerel yönetimi temsil eden kategorik ID. Ortalama: 347, Max: 941 → sayısal analiz uygun değil. Sayı gibi değil, etiket gibi işlenmeli.
- 1st_Road_Class: 1–6 arasında değer alan ordinal kategorik değişken. Ortalama ≈ Medyan → simetrik dağılım. Anlamlı sıralama olabilir (örnek: otoyol → tali yol) → ordinal encoding önerilir.
- 1st_Road_Number: Sayısal bir değişken gibi gözüküyor, ancak -1 değeri mevcut → eksik veri kodlaması olabilir. Max 9999 → uç değer bariz, median sadece 129. -1'ler NaN yapılmalı, uç değerler analiz edilerek filtrelenmeli.
- Speed_limit: Sürekli sayısal değişken. Q1 = 30, Q3 = 50 → standart hız limitleri aralığında. Mean = 39 → hafif pozitif çarpıklık var ama genel olarak temiz. Modellemeye direkt dahil edilebilir.
- 2nd_Road_Class: -1 değeri çok fazla → eksik veri içeren ordinal kategorik değişken. Ortalama = 2.67, medyan = 3 → dağılım dengesiz. -1 → NaN yapılmalı, sonra kategori encoding yapılmalı.
- 2nd_Road_Number: -1 değeri mevcut → eksik veri kodlaması olabilir. Medyan = 0, ama max 9999 → pozitif çarpık, uç değerli. -1'ler NaN yapılmalı, dağılım normalize edilmeli ya da binleme yapılabilir.
- Urban_or_Rural_Area: 1, 2, 3 gibi kodlanmış kategorik değişken (1: Urban, 2: Rural, 3: bilinmiyor olabilir). Ortalama = 1.35 → çoğunlukla şehir kazaları. Dengesiz kategori olabilir → oranlarına göre sınıflandırmada dikkatli davranılmalı.

```python
import matplotlib.pyplot as plt

gender_counts = df["Accident_Severity"].value_counts()
plt.figure(figsize=(6, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Trafik Kazalarında Ciddiyet Dağılımı")
plt.ylabel("")  # y-axis etiketini kaldırmak için
plt.show()
```

Grafiğe baktığımızda büyük bir çoğunluğunun 3.derece olduğunu çok küçük bir dilimin 1.derece kazalardan oluştuğunu söyleyebiliriz.

```python
import seaborn as sns
# 'Pedestrian_Crossing-Human_Control' sütunundaki değerlerin sayısı
satisfaction_counts = df["Pedestrian_Crossing-Human_Control"].value_counts()

plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x="Pedestrian_Crossing-Human_Control", palette="coolwarm")

plt.title("Yaya Geçidi İnsan Kontrolü Dağılımı", fontsize=16)
plt.xlabel("Kontrol Tipi", fontsize=12)
plt.ylabel("Kaza Sayısı", fontsize=12)

plt.xticks(rotation=45)  # Eğer kategoriler uzun ise döndürür
plt.tight_layout()       # Grafik elemanlarının taşmaması için

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 5,  # biraz üstünde yazması için +5 ekledim
            int(height), ha="center", fontsize=10)
    
plt.show()
```


Bu grafik, kazaların oluştuğu noktalarda kontrol noktalarının önemini çarpıcı bir şekilde ortaya koyuyor. Kazaların büyük bir çoğunluğu, olay yerinin 50 metre yakınında herhangi bir kontrol noktası bulunmadığında meydana gelmiş.
